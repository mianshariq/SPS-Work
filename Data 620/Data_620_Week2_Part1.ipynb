# Getting Data From the Web

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import json
import csv
```

```python
nodes=pd.read_csv('https://github.com/mianshariq/Data-Science/raw/865c0a3fde79f1ccf6caf019852dd62be9206850/SC_Nodes.csv')
edges=pd.read_csv('https://github.com/mianshariq/Data-Science/raw/865c0a3fde79f1ccf6caf019852dd62be9206850/SC_edges.csv')
```

```python
nodesJ=nodes.to_json(orient='records')
nodesJ = json.loads(nodesJ)
edgesJ=edges.to_json(orient='records')
edgesJ = json.loads(edgesJ)
```

```python
import nltk
import matplotlib.pyplot as plt


# Load the Brown corpus
from nltk.corpus import brown
import nltk

nltk.download('brown')

# Get the most common words in the corpus
words = brown.words()
fdist = nltk.FreqDist(words)
most_common_words = [word for word, _ in fdist.most_common()]


```

```python
import numpy as np
```

# 1. Calculate the number of unique words representing half of the total words


```python
half_total_words = len(words) // 2
unique_words = set()

for word in most_common_words:
    unique_words.add(word)
    if len(unique_words) >= half_total_words:
        break

num_unique_words = len(unique_words)
print(f"Number of unique words representing half of the total words: {num_unique_words}")
```

# 2. Get the 200 highest frequency words in the corpus

```python
top_200_words = most_common_words[:200]
print(top_200_words)
```


# 3. Create a graph showing the relative frequency of these 200 words

```python
x = range(1, 201)
y = [fdist.freq(word) for word in top_200_words]

plt.plot(x, y)
plt.xlabel("Rank")
plt.ylabel("Relative Frequency")
plt.title("Relative Frequency of Top 200 Words")
plt.show()
```

# 4. Does the observed relative frequency of these words follow Zipfâ€™s law? Explain.
# Zipf's law states that the frequency of a word is inversely proportional to its rank.
# We can check this by plotting the log of rank against the log of frequency and examining the slope of the line.

```python
log_rank = [np.log(i) for i in x]
log_frequency = [np.log(fdist.freq(word)) for word in top_200_words]

plt.plot(log_rank, log_frequency)
plt.xlabel("Log Rank")
plt.ylabel("Log Frequency")
plt.title("Zipf's Law: Log Rank vs. Log Frequency")
plt.show()
```
